Convert all letters of the given string expression to uppercase. Put space instead of comma and point, separate word by word.

```python
text = "The goal is to turn data into information, and information into insight."
text = text.replace(",", " ").replace(".", " ")
text = text.split()
text = [word.upper() for word in text]
print(text)
```

Write a function that takes a list as an argument, assigns the odd and even numbers in the list to separate lists, and returns these lists.

```python
l = [2,13,18,93,22]

sayi = [[],[]]
def odd_even(l):
    for number in l:
        if number % 2 == 0:
            sayi[0].append(number)
        else:
            sayi[1].append(number)
    return sayi

odd_even(l)
```

The list given below contains the names of the students who were successful in the engineering and medical faculties. While the first three students respectively represent the success rank of the faculty of engineering, the last three students belong to the student rank of the faculty of medicine. Print student degrees by faculty using Enumarate

```python
ogrenciler = ["Ali","Veli","Ayşe","Talat","Zeynep","Ece"]

for index, ogrenci in enumerate(ogrenciler, 1):
    if index < 4:
        print(f"Mühendislik Fakültesi: {index}. ogrencisi: {ogrenci}")
    else:
        print(f"Tıp Fakültesi: {index}. ogrencisi: {ogrenci}")
```

Three lists are given below. The lists include course code, credit and quota information, respectively. Print the course information using Zip.

```python
ders_kodlari = ["CMP1005","PSY1001","HUK1005","SEN2204"]
krediler = [3,4,2,4]
kontenjanlar = [30,75,150,25]

for derskodu, kredi, kontenjan in zip(ders_kodlari,krediler,kontenjanlar):
    print(f"Kredisi {kredi} olan {derskodu} kodlu dersin kontenjani {kontenjan} kisidir.")
```
